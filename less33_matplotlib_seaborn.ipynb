```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('cars.xlsx')
data
```

```python
sorted_data = data.sort_values(by='Споживання палива (л/100км)')
plt.plot(sorted_data['Споживання палива (л/100км)'], sorted_data['Вартість (у доларах)'], color='blue')
plt.xlabel('Споживання палива (л/100км)', color='blue')
plt.ylabel('Вартість (у доларах)', color='blue')
plt.title('Залежність споживання палива від вартості', color='blue')
plt.show()
```

```python
С = data['Споживання палива (л/100км)']
A = data['Максимальна швидкість (км/год)']
plt.scatter(С, A, color = 'red')
plt.xlabel('Споживання палива (л/100км)',color = 'red')
plt.ylabel('Максимальна швидкість (км/год)', color = 'red')
plt.title('Залежність споживання палива від максимальної швидкості',color='red')
plt.show()
```

```python
B = data['Вантажопідйомність (кг)']
plt.scatter(B,A,color = 'green')
plt.xlabel('Вантажопідйомність (кг)',color = 'green')
plt.ylabel('Максимальна швидкість (км/год)', color = 'green')
plt.title('Залежність вантажопідйомності від макс. швидкості', color = 'green')
plt.show()
```

```python
#для розподілу авто на 5 груп зробив 2 графіки, бо не ясно за якими критеріями та у якому вигляді повинен бути цей розподіл (хоча цей перший графік не зовсім гістограма)
data['Швидкість'] = pd.cut(data['Максимальна швидкість (км/год)'], bins=5)
data = data.sort_values(by=['Марка авто'])

g = sns.barplot(x='Марка авто', y='Максимальна швидкість (км/год)', data=data, hue='Швидкість', dodge=False)
g.set(ylabel='Максимальна швидкість (км/год)')
g.set_xticklabels(g.get_xticklabels(), rotation=90)
g.legend(fontsize=6)
g.legend(title='Групи авто (швидкість)', loc='upper right', bbox_to_anchor=(1.3, 1))
g.figure.set_size_inches(12, 6)
plt.yticks(range(0, 280, 10))
plt.title('Розподіл авто на 5 груп по швидкості')
plt.show()
```

```python
data['Швидкість'] = pd.cut(data['Максимальна швидкість (км/год)'], bins=5)
sns.histplot(data=data, x='Максимальна швидкість (км/год)', hue='Швидкість', multiple='stack', bins=10).set_ylabel('Кількість авто')
plt.title('Розподіл авто на 5 груп по швидкості')
plt.show()
```

```python
b = sns.boxplot(x='Марка авто', y='Максимальна швидкість (км/год)', data=data)
b.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.yticks(range(160, 261, 10))
plt.title('Коробковий графік для відображення швидкості різних авто')
plt.show()
```
